```python
import numpy as np
data = np.genfromtxt('/content/1774404110.csv', delimiter=',', dtype=None, encoding=None, names=True)
mpg = data['mpg']
cylinders = data['cylinders']
displacement = data['displacement']
horsepower = data['horsepower']
origin = data['origin']
car_name = data['car_name']
```

```python
mean_mpg = np.mean(mpg)
median_mpg = np.median(mpg)
std_mpg = np.std(mpg)

print("Mean:", mean_mpg)
print("Median:", median_mpg)
print("Standard Deviation:", std_mpg)
```

```python
count_high_mpg = np.sum(mpg > 25)
print("Cars with mpg > 25:", count_high_mpg)
```

```python
filtered_cars = car_name[cylinders > 6]
filtered_list = filtered_cars.tolist()

print(filtered_list)
```

```python
import numpy as np
horsepower_clean = np.array([
    float(x) if x != '?' else np.nan
    for x in horsepower
])
hp_mean = np.nanmean(horsepower_clean)
horsepower_clean[np.isnan(horsepower_clean)] = hp_mean
hp_updated = horsepower_clean * 1.10

print("Updated Horsepower:")
print(hp_updated[:10])
```

```python
europe_disp = displacement[origin == 2]
avg_disp = np.mean(europe_disp)

print("Average Displacement (Europe):", avg_disp)
```

```python
unique_cyl = np.unique(cylinders)
print("Mean MPG for each cylinder group:")
for cyl in unique_cyl:
    avg_mpg = np.mean(mpg[cylinders == cyl])
    print(f"Cylinders {cyl}: {avg_mpg}")
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/1774404110.csv")
print("First 10 rows:")
print(df.head(10))

print("\nShape of dataset (rows, columns):")
print(df.shape)

print("\nSummary statistics:")
print(df.describe())
```

```python
filtered_df = df[(df['model year'] == 75) & (df['weight'] < 3000)]

result = filtered_df[['car name', 'weight', 'mpg']]

print(result)
```

```python
print("\nMissing values:")
print(df.isnull().sum())
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
median_hp = df['horsepower'].median()
df['horsepower'].fillna(median_hp, inplace=True)
```

```python
# 4. Data Transformation
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']

print("\nNew column added (power_to_weight_ratio):")
print(df[['horsepower', 'weight', 'power_to_weight_ratio']].head())

```

```python
# --------------------------------------------------
# 5. Group By
grouped = df.groupby('origin')['mpg'].mean()

print("\nMean MPG by Origin:")
print(grouped)
```

```python
summary_df = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()

print("\nSummary DataFrame (by model year):")
print(summary_df)
```

```python
# 7. Exporting Data
high_mpg = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]

high_mpg.to_csv("high_mpg_cars.csv", index=False)

print("\nFile 'high_mpg_cars.csv' created successfully!")
```
