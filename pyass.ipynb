```python
FILE_NAME = "expenses.txt"
def add_expense():
    category = input("Enter category (Food/Travel/etc.): ")

    try:
        amount = float(input("Enter amount: "))
    except ValueError:
        print("Invalid amount! Please enter a number.")
        return

    try:
        with open(FILE_NAME, "a") as file:
            file.write(f"{category},{amount}\n")
        print("Expense added successfully!")
    except Exception as e:
        print("Error writing to file:", e)



def view_expenses():
    try:
        with open(FILE_NAME, "r") as file:
            print("\n--- All Expenses ---")
            for line in file:
                category, amount = line.strip().split(",")
                print(f"Category: {category} | Amount: ₹{amount}")
    except FileNotFoundError:
        print("No expenses found. File does not exist yet.")



def calculate_total():
    total = 0
    try:
        with open(FILE_NAME, "r") as file:
            for line in file:
                _, amount = line.strip().split(",")
                total += float(amount)
        print(f"\n Total Expense: ₹{total}")
    except FileNotFoundError:
        print(" No data to calculate. File not found.")



def category_count():
    counts = {}
    try:
        with open(FILE_NAME, "r") as file:
            for line in file:
                category, _ = line.strip().split(",")
                counts[category] = counts.get(category, 0) + 1

        print("\n Category-wise Count:")
        for category, count in counts.items():
            print(f"{category}: {count}")

    except FileNotFoundError:
        print(" No data available. File not found.")



def main():
    while True:
        print("\n==== Expense Tracker ====")
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Total Expense")
        print("4. Category-wise Count")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_expense()
        elif choice == "2":
            view_expenses()
        elif choice == "3":
            calculate_total()
        elif choice == "4":
            category_count()
        elif choice == "5":
            print(" Exiting program...")
            break
        else:
            print(" Invalid choice. Try again.")



if __name__ == "__main__":
    main()
```
